```python
import gym
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python

environment = gym.make("FrozenLake-v1", is_slippery=False)
environment.reset()
environment.render()
```

```python
# tablomuz ilk ba≈üta 0 larla doludur √ß√ºnk√º daha √∂ƒürenme yapƒ±lmamƒ±≈ütƒ±r
#(buluduƒüu konum x yapƒ±lan eylem) se≈ükinde tablo olu≈üur
qtable = np.zeros((16, 4))

nb_states = environment.observation_space.n  # 16
nb_actions = environment.action_space.n     # 4
qtable = np.zeros((nb_states, nb_actions))

# tablomuzu yazdƒ±rƒ±p, bakalƒ±m.
print("Q-table = ")
print(qtable)
```

```python
"""
‚óÄÔ∏è SOL = 0 
üîΩ A≈ûAƒûI = 1 
‚óÄÔ∏è SAƒû = 2 
üîº YUKARI = 3
"""
```

```python
action = environment.action_space.sample()

new_state, reward, done, truncated, info = environment.step(action)

environment.render()
print(f"Reward = {reward}")
print(action)
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams.update({'font.size': 15})

# Q tobosunu ba≈ülatƒ±yoruz
qtable = np.zeros((environment.observation_space.n, environment.action_space.n))

# Parametrelerimiz
episodes =100         # Ajanƒ±n ka√ß kez √ßevre ile etkile≈üime girecegi. arttƒ±kca deneyim artacaƒüƒ±ndan kazanma olasƒ±lƒ±ƒüƒ±da artar.
alpha = 0.5            # Ajanƒ±n √∂ƒürenme hƒ±zƒ±. √ßok y√ºksek √∂ƒürenme oranƒ± √∂nceki deneyimleri unutmasƒ±na yol a√ßar
gamma = 0.9            # indirim fakt√∂r√º: ajanƒ±n gelecekteki √∂d√ºlleri ne kadar √∂nemsediƒüini kontrl eder

# Sonu√ßlarƒ±n listesi
outcomes = []

# 0 larla dolu eski Q toblomuzu yazdƒ±rƒ±r
print('√ñƒürenim √∂ncesi Q tablosu:')
print(qtable)

# Training-Q learning

# yukarƒ±da tanƒ±mlanan episodes deƒüeri kadar ajan deneme hakkƒ± verir
for _ in range(episodes):
    state = (environment.reset()[0])      # √ßevreyi sƒ±fƒ±rlar

    done = False

    # sonu√ß ba≈üarƒ±sƒ±zlƒ±k olarak alƒ±nƒ±r
    outcomes.append("Failure")

    # ajan deliƒüe d√º≈ümdeiƒüi s√ºrece √∂ƒürenmeye devam eder
    while not done:
        
        # i√ßinde bulunduƒüu durumda en y√ºksek deƒüere sahip olan eylemi se√ß√ß.
        if np.max(qtable[state]) > 0:
          action = np.argmax(qtable[state])

        # en iyi eylem yoksa, b√ºt√ºn deƒüerle e≈üitse random se√ßim yap
        else:
          action = environment.action_space.sample()
             
        # Implement this action and move the agent in the desired direction
        new_state, reward, done,_, info = environment.step(action)

        # Q tablosundaki Q(s,a) deƒüeirni form√ºle g√∂re g√ºnceller
        qtable[state, action] = qtable[state, action] + \
                                alpha * (reward + gamma * np.max(qtable[new_state]) - qtable[state, action])
        
        # mecut yeni konumu g√ºncelle 
        state = new_state

        # $$$$ tek √∂de√ºl hedefte olduƒüu i√ßin , √∂d√ºl varsa hedefe ula≈üƒ±lmƒ±≈ü demektir
        if reward:
          outcomes[-1] = "Success"
          
#print()
print('===========================================')
print('√ñƒürenim sonrasƒ± Q tablosu:')
print(qtable)

# Tablo √∂zellikleri
plt.figure(figsize=(12, 5))
plt.xlabel("Run number")
plt.ylabel("Outcome")
plt.plot(range(len(outcomes)), outcomes, color="red", marker='o')
plt.show()
 
plt.show()


# ba≈üƒ±na $$$$ koyduƒüum yerd3e outcomes listesine Success koyuyoruz. count() fonksiyonu ile bu sayƒ±yƒ± buluyorxu)
ba≈üarƒ±_y√ºzde = outcomes.count("Success")  # Ba≈üarƒ±lƒ± denemelerin sayƒ±sƒ±

total_episodes = len(outcomes)  # Toplam deneme sayƒ±sƒ±

hesaplama = (ba≈üarƒ±_y√ºzde / total_episodes) * 100  # Ba≈üarƒ± y√ºzdesi hesapla
hesaplama= int(hesaplama)
print(f"Ba≈üarƒ±: % {hesaplama}")




```
